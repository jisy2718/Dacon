### 2-1. 여러 모델 비교
I want to compare other models (Logistic Regression, Linear SVM, Random Forest , etc) by validation f1 score.
Then you choose best model, and make submission_bestmodel.csv from selected model.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load the training data
train_data = pd.read_csv('train.csv')

# Load the testing data
test_data = pd.read_csv('test.csv')

# Split the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(train_data['text'], train_data['label'], test_size=0.2, random_state=42)

# Encode the labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)

# Define the models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Linear SVM': LinearSVC(),
    'Random Forest': RandomForestClassifier(),
    'Multinomial Naive Bayes': MultinomialNB()
}

# Train and validate each model, and store the F1 scores
f1_scores = {}
for model_name, model in models.items():
    pipeline = Pipeline([
        ('tfidf', TfidfVectorizer()),
        ('classifier', model)
    ])
    pipeline.fit(X_train, y_train_encoded)
    y_val_pred = pipeline.predict(X_val)
    f1 = f1_score(y_val_encoded, y_val_pred, average='weighted')
    f1_scores[model_name] = f1
    print(f"{model_name} F1 score: {f1:.3f}")

# Choose the best model
best_model_name, best_f1_score = max(f1_scores.items(), key=lambda x: x[1])
print(f"Best model: {best_model_name} with F1 score: {best_f1_score:.3f}")

# Train the best model on the whole training set
best_model = models[best_model_name]
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('classifier', best_model)
])
pipeline.fit(train_data['text'], label_encoder.transform(train_data['label']))

# Predict the categories for the test set
y_test_pred = pipeline.predict(test_data['text'])
test_data['label'] = label_encoder.inverse_transform(y_test_pred)

# Save the submission file
submission =test_data[['id', 'label']]
submission.to_csv('submission_bestmodel.csv', index=False)
```
